## Various DTypes and distributions

Plots are automatically chosen based on the data type and the distribution of the data.

| DType  | Size | Type of plot
| --- | --- | --- |
| numerical | small | bar |
| numerical | big   | histogram |
| object | small / big | bar |
| timeseries  | small / big | line |
